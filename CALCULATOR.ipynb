```python
from tkinter import *

window = Tk()
window.title('Simple Calculator')
window.geometry('300x275')

def click_button(value):
    current = expression_field.get()
    expression_field.delete(0, END)
    expression_field.insert(0, str(current) + str(value))
    
    
def evaluate_expression():
    try:
        expression = expression_field.get()
        result = eval(expression)
        expression_field.delete(0, END)
        expression_field.insert(0, str(result))
    except Exception as e:
        expression_field.delete(0, END)
        expression_field.insert(0, "Error")
        

def clear_expression():
    expression_field.delete(0, END)

    
expression_field = Entry(window, width=16, font=('Arial', 24))  
expression_field.grid(columnspan=5)


B1 = Button(window, text="1", bg="white", fg="black", command=lambda: click_button(1),width=5,font=("Arial",14))
B1.grid(row=1, column=1) 
B2 = Button(window, text="2", bg="white", fg="black", command=lambda: click_button(2),width=5,font=("Arial",14))
B2.grid(row=1, column=2) 
B3 = Button(window, text="3", bg="white", fg="black", command=lambda: click_button(3),width=5,font=("Arial",14))
B3.grid(row=1, column=3)  
B4 = Button(window, text="4", bg="white", fg="black", command=lambda: click_button(4),width=5,font=("Arial",14),)
B4.grid(row=2, column=1)
B5 = Button(window, text="5", bg="white", fg="black", command=lambda: click_button(5),width=5,font=("Arial",14))
B5.grid(row=2, column=2)
B6 = Button(window, text="6", bg="white", fg="black", command=lambda: click_button(6),width=5,font=("Arial",14))
B6.grid(row=2, column=3)
B7 = Button(window, text="7", bg="white", fg="black", command=lambda: click_button(7),width=5,font=("Arial",14))
B7.grid(row=3, column=1)
B8 = Button(window, text="8", bg="white", fg="black", command=lambda: click_button(8),width=5,font=("Arial",14))
B8.grid(row=3, column=2)
B9 = Button(window, text="9", bg="white", fg="black", command=lambda: click_button(9),width=5,font=("Arial",14))
B9.grid(row=3, column=3) 
B0 = Button(window, text="0", bg="white", fg="black", command=lambda: click_button(0),width=5,font=("Arial",14))
B0.grid(row=4, column=2) 
Bplus = Button(window, text="+", bg="white", fg="black", command=lambda: click_button("+"),width=5,font=("Arial",14))
Bplus.grid(row=1, column=4)
Bminus = Button(window, text="-", bg="white", fg="black", command=lambda: click_button("-"),width=5,font=("Arial",14))
Bminus.grid(row=2, column=4) 
Bmul = Button(window, text="*", bg="white", fg="black", command=lambda: click_button("*"),width=5,font=("Arial",14))
Bmul.grid(row=3, column=4) 
Bdiv = Button(window, text="/", bg="white", fg="black", command=lambda: click_button("/"),width=5,font=("Arial",14))
Bdiv.grid(row=4, column=4) 
Bopen = Button(window, text="(", bg="white", fg="black", command=lambda: click_button("("),width=5,font=("Arial",14))
Bopen.grid(row=4, column=1) 
Bclose = Button(window, text=")", bg="white", fg="black", command=lambda: click_button(")"),width=5,font=("Arial",14))
Bclose.grid(row=4, column=3) 
Bequal = Button(window, text="=", bg="white", fg="black", command=evaluate_expression,width=11,font=("Arial",14))
Bequal.grid(row=5, column=3,columnspan=2) 
Bclear = Button(window, text="C", bg="white", fg="black", command=clear_expression,width=11,font=("Arial",14))
Bclear.grid(row=5, column=1,columnspan=2) 

window.mainloop()
```
